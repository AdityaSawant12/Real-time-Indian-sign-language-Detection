```python
import cv2
import os

# Function to create the directory if it does not exist
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Set directory to save images
image_directory = 'Data/26'  # Change '0' to any label class folder if needed
create_directory(image_directory)

# Initialize the camera
cap = cv2.VideoCapture(0)

# Set image counter
img_counter = 0

print("Press 'q' to quit...")

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Display the live video feed
    cv2.imshow("Press 'q' to capture", frame)

    # Save image every frame (customize when needed, here it's frame by frame)
    img_name = os.path.join(image_directory, f"{img_counter}.jpg")
    cv2.imwrite(img_name, frame)
    print(f"{img_name} saved!")
    img_counter += 1

    # If 'q' is pressed, break the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Stopping the capture...")
        break

# Release the capture and destroy windows
cap.release()
cv2.destroyAllWindows()
```
